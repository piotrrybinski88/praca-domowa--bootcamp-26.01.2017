```python
# sieve returns boolean that number is Prime.

def czyPierwsze(array):
    """
    argument:  one dimensional array of integers
    return: boolean array where checked the number is prime or not  
    """
    return np.where([num in sieve for num in array] , "True", "False")


# test :
array1 = np.array([1, 10, 3, 5, 12, 13, 27, 37])
print(czyPierwsze(array1))
print(czyPierwsze(np.array([1,2,3,4,5,6,7,11,22,19])))
print(czyPierwsze(np.array([123, 234, 123525, 443, 157])))
```

```python
def isFirst(array):
    """
    argument:  one dimensional array of integers
    return: boolean array where checked the number is prime or not  
    """
    def isPrime(num):
        
        if num == 1:
            return False
        elif num > 1:
            for i in range(2, int(np.trunc(num ** 0.5))):  # zmniejsza ilość liczb przez, które dzielimy
                if num % i == 0:
                    return False
            else:
                return True 

    return [isPrime(i) for i in array]
```

```python
print(isFirst([1,2,3, 15, 11, 12, 234, 221 , 17]))
print(isFirst(array1))
```

###### Write dwumian(n, k) - binominal coefficient n of k

```python

def halfFactorial(y, x = 1):
    """return - product of numbers from x to y  """
    result = 1
    
    if x == 0:
        x = 1

    for i in range(x, y + 1):
        result *= i
        
    return result


def coefNewton(n, k):
    """
    returns binomial / Newton coefficient of n and k, where n and k must be integer
    and must follow those rules n >= k >= 0
    
    """
    if not type(n) == int or not type(n) == int:
        raise ValueError("n and k must be integer")
        
    elif not n >= k >= 0:
        raise ValueError("n and k must follow those rules n >= k >= 0")
        
    elif (n - k) >= k:
        up = halfFactorial(x = n - k + 1, y = n)
        down = halfFactorial(x = 1,y =  k )
        
    
    else:
        up = halfFactorial(x = k + 1, y =  n)
        down = halfFactorial(n - k )
        
    coef = up / down 
    
    return coef
        

print(coefNewton(10, 2))
print(coefNewton(10, 8))
print(coefNewton(10,10))
print(coefNewton(10,1))
print(coefNewton(28,25))
print(coefNewton(20,18))
print(coefNewton(30, 22))
print(coefNewton(3.2,3))
```

###### Zadanie 2.4

```python
def wartoscWielomianu(wspolczyniki, x):
    """ argumenty - wektor współczyników, x - punkt dla którego obliczamy wartość """
    potega = np.arange(1, len(wspolczyniki))
    wartosc = wspolczyniki[0]  + np.sum(wspolczyniki[1:] * x ** potega)
    
    return wartosc 

a = np.array([1, 1, 1])
b = np.array([1, 2, 3])
c = np.array([1, 2, 3])
d = np.array([1, 2, 3, 4, 5])

print(wartoscWielomianu(a, 0))
print(wartoscWielomianu(b, 1))
print(wartoscWielomianu(c, 2))
print(wartoscWielomianu(d, 2))
```
