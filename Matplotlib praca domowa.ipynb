```python

def half_factorial(y, x = 1):
    """return - product of numbers from x to y  """
    result = 1
    
    if x == 0:
        x = 1

    for i in range(x, y + 1):
        result *= i
        
    return result


def coefNewton(n, k):
    """
    returns binomial / Newton coefficient of n and k, where n and k must be integer
    and must follow those rules n >= k >= 0
    
    """
    if not type(n) == int or not type(n) == int:
        raise ValueError("n and k must be integer")
        
    elif not n >= k >= 0:
        raise ValueError("n and k must follow those rules n >= k >= 0")
        
    elif (n - k) >= k:
        up = half_factorial(x = n - k + 1, y = n)
        down = half_factorial(x = 1,y =  k )
        
    
    else:
        up = half_factorial(x = k + 1, y =  n)
        down = half_factorial(n - k )
        
    coef = up / down 
    
    return coef
        

print(coefNewton(10, 2))
print(coefNewton(10, 8))
print(coefNewton(10,10))
print(coefNewton(10,1))
print(coefNewton(28,25))
print(coefNewton(20,18))
print(coefNewton(30, 22))
print(coefNewton(3.2,3))
```

```python
def factorial(y, x = 1):
    wynik = 1
    if x == 0:
        return wynik

    for i in range(x, y):
        wynik *= i 
    return wynik

print(factorial(y = 30, x = 23))


```
